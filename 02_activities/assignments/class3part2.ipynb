```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import PIL
import requests
import scipy

from matplotlib.ticker import MaxNLocator
```

```python
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Nami","Usoop", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# fig, (ax1, ax2) = plt.subplots(ncols = 2,
#                                nrows = 1,
#                                figsize = (7, 3))

# ax1.scatter (x1, y1)
# ax2.bar(x2, y2)



# Create figure with subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(12, 5))
# fig.suptitle('One Piece Character Analysis', fontsize=14)

# Customize scatter plot
ax1.scatter(x1, y1, 
            c='navy', 
            s=50,
            alpha=0.7,
            marker='o')

ax1.set_title('Scatter Plot', fontsize=12)
ax1.set_xlabel('Training Days')
ax1.set_ylabel('Power Level')
ax1.grid(True, linestyle='--', alpha=0.5)

# Customize bar plot
colors = ['#FF9933', '#77B300', '#33CCFF', '#CC6600', '#FF3399']
bars = ax2.bar(x2, y2, 
        color=colors,
        width=0.6,
        edgecolor='black')

ax2.set_title('Bar Graph', fontsize=12)
ax2.set_xlabel('Character')
ax2.set_ylabel('Strength Level')

# Add values on top of bars
for bar in bars:
    height = bar.get_height()
    ax2.annotate(f'{height}',
                xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3),
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7, 4),
                                    layout="constrained")
#   layout="constrained" - ensure nothing is outside the figure


someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1, y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize=14)
someaxes['ax2'].set_xlabel('Another label', fontsize=10)
someaxes['ax3'].set_xlabel('Label 2: Fast and Furious', fontsize=12)
```

```python
x = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])

fig, ax = plt.subplots(figsize=(7,3))

ax.bar(x,y1,color = 'indigo')
ax.plot(x,y2,color = 'red')
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2,color = 'red')

ax.errorbar(x,y2,
            yerr=y2_sd,
            fmt='none',
            elinewidth=4,
            capsize=6,
            capthick=2,
            errorevery=2)

#    errorevery=2. not super useful. - displays error bars every 2nd data point
```
