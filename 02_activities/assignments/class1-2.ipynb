```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613)
#random seed ensures that every time we run the code, we get the same sequence of random numbers. 
#useful for reproducibility.

x = np.arange(50)
# creates an array of numbers from 0 to 49

y = np.random.randint(0, 100, 50)
# generates an array of 50 random integers between 0 and 99

fig, ax = plt.subplots(figsize = (5,3))
# creates a figure (fig) and a set of axes (ax).
# sets the figure size to 5 inches wide and 3 inches tall.


# creating various types of graphs:

# ax.scatter(x, y)
# creates scatter plot, where each (x, y) pair is plotted as a point.

# ax.bar(x,y)
# creates bar chart, where each x value has a corresponding bar of height y.

# ax.plot(x,y)
# create a line plot, connecting points (x, y) with lines.

ax.hist(y)
# creates a histogram of y: groups the values in y into bins and counts how many values fall into each bin.
```

```python
# modifying labels and titles

# define font properties for the plot labels and title,
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

# create a figure (fig) and a set of axes (ax). 
# set the figure size to 5 inches wide and 3 inches tall.

fig, ax = plt.subplots(figsize = (5,3))
# plot x (years) against y (growth values) using a line plot.
ax.plot(x,y)

# set titles for a pot, y- and x-asis labels. sets fonts/styles
ax.set_title('Total growth overtime', fontdict=font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict=font2, loc = 'bottom')
ax.set_xlabel('Years since start', fontdict=font2)
```

```python
# Libraries
# import numpy as np
# import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
 
# Create data: 200 points
data = np.random.multivariate_normal([0, 0], [[1, 0.5], [0.5, 3]], 200)
x, y = data.T
 
# Create a figure with 6 plot areas
fig, axes = plt.subplots(ncols=7, nrows=1, figsize=(25, 5))
 
# Everything starts with a Scatterplot
axes[0].set_title('Scatterplot')
axes[0].plot(x, y, 'ko')
# As you can see there is a lot of overlapping here!

# Everything starts with a Scatterplot
axes[1].set_title('Scatterplot')
axes[1].scatter(x, y, color='green', edgecolor='black')
 
# # Thus we can cut the plotting window in several hexbins
nbins = 20
axes[2].set_title('Hexbin')
axes[2].hexbin(x, y, gridsize=nbins, cmap=plt.cm.BuGn_r)
 
# # 2D Histogram
axes[3].set_title('2D Histogram')
axes[3].hist2d(x, y, bins=nbins, cmap=plt.cm.BuGn_r)
 
# Evaluate a gaussian kde on a regular grid of nbins x nbins over data extents
k = gaussian_kde(data.T)
xi, yi = np.mgrid[x.min():x.max():nbins*1j, y.min():y.max():nbins*1j]
zi = k(np.vstack([xi.flatten(), yi.flatten()]))
 
# plot a density
axes[4].set_title('Calculate Gaussian KDE')
axes[4].pcolormesh(xi, yi, zi.reshape(xi.shape), shading='auto', cmap=plt.cm.BuGn_r)
 
# add shading
axes[5].set_title('2D Density with shading')
axes[5].pcolormesh(xi, yi, zi.reshape(xi.shape), shading='gouraud', cmap=plt.cm.BuGn_r)
 
# contour
axes[6].set_title('Contour')
axes[6].pcolormesh(xi, yi, zi.reshape(xi.shape), shading='gouraud', cmap=plt.cm.BuGn_r)
axes[6].contour(xi, yi, zi.reshape(xi.shape) )

plt.show()
```
